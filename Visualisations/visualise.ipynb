```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns

%matplotlib inline 
```

```python
df=pd.read_json('C:\\Cricket\\Data\\playerStats.json')
df.shape
```

```python
batting_df=df.iloc[:,[0,1,2]]
bowling_df=df.iloc[:,[0,3,4]]
```

```python
bowling_df.shape
batting_df.shape
```

```python
batting_df=batting_df.drop(batting_df[batting_df.Runs == 0].index)
bowling_df=bowling_df.drop(bowling_df[bowling_df.Wickets == 0].index)
bowling_df=bowling_df.drop(bowling_df[bowling_df.OversBowled < 1].index)
```

```python
sns.catplot(batting_df,kind='bar', x='BallsFaced' , y='Runs')
```

```python
batting_df['strikeRate']= (batting_df['Runs'] / batting_df['BallsFaced'] )*100
```

```python
sns.catplot(batting_df,x='Runs',y='strikeRate' , kind='bar')
```

```python
bowling_df['strikeRate']= (bowling_df['Wickets'] / bowling_df['OversBowled'])*100
```

```python
sns.catplot(bowling_df,kind='bar',x='strikeRate',y='Wickets')
```
